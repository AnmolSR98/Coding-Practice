```python
#question 1, checking merely if they are equal

gr1 = input("What is the first grade? ")
gr2 = input("What is the second grade? 3")

if gr1 == gr2:
  print(gr1 + " and " + gr2 + " are equal!")

else:
  print(gr1 + " and " + gr2 + " are not equal!")
```

```python
#question 2, checking which one is the highest

gr1 = input("What is the first grade? ")
gr2 = input("What is the second grade? ")
gr3 = input("What is the third grade? ")

if gr1 == gr2 == gr3:
  print("They're all equal!")

else:
  if gr1 > gr2 and gr1 > gr3:
    print(gr1 + " is the largest grade!")
  elif gr2 > gr1 and gr2 > gr3:
    print(gr2 + " is the largest grade!")
  else:
    print(gr3 + " is the largest grade!")
```

```python
#question 3, marking each as failure, poor, satisfactory, good, excellent, and invalid

gr = float(input("What is the grade? "))

st = "The grade is"

if gr >= 0 and gr < 1:
  print(st + " a failing grade.")

elif gr >= 1 and gr < 1.3:
  print(st + " a poor grade.") 

elif gr >= 1.3 and gr < 2.3:
  print(st + " a satisfactory grade.") 

elif gr >= 2.3 and gr < 3.3:
  print(st + " a good grade.") 

elif gr >= 3.3 and gr <= 3.3:
  print(st + " an excellent grade.") 

else:
  print("Please restart the program and enter a valid grade of [0,4], element the reals.")
```

```python
#question 4, checking which level to assign them to

gr1 = float(input("What is the score on quiz #1? "))
gr2 = float(input("What is the score on quiz #2? "))

grarr = [gr1, gr2]

st = "You have been selected for level"

if min(grarr) >= 80:
  print(st + " 3.")

elif min(grarr) >= 50:
  print(st + " 2.")

elif min(grarr) <= 50 and max(grarr) >= 50:
  print(st + " 2 but will have to redo the quiz on which you received a(n)", min(grarr))

else:
  print(st + " 1.")
```
